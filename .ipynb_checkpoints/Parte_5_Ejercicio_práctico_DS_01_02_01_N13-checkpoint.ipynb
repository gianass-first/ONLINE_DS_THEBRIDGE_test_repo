## Ejercicio 10
Convierte el programa del ejercicio 2 en una función al que se le pase el valor a buscar como un argumento posicional

`````python
def buscar_indices(lista, palabra):
    indices = []
    
    for i in range(len(lista)):
        if lista[i].lower() == palabra.lower():
            indices.append(i)
    
    return indices

ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

print(buscar_indices(ejer_3, "un"))
print(buscar_indices(ejer_3, "es"))
print(buscar_indices(ejer_3, "binario"))
`````

## Ejercicio 11
Convierte el programa del ejercicio 9 en una función al que se le pasen como argumentos la librería, posicional, y el título, argumento keyword con valor "ninguno", por defecto.

`````python
def buscar_libros(libreria, titulo="ninguno"):

    for i in range(len(libreria)):
        if libreria[i]["titulo"].lower() == titulo.lower():
            return "Encontrado"
    return "Ay mi madre el bicho, no lo tengo"

print(buscar_libros(libreria, "el señor de los anillos"))
print(buscar_libros(libreria, "Don Quijote"))
`````

## Ejercicio 12

Solicitar al usuario que ingrese su dirección email. Imprimir un mensaje indicando si la dirección es válida o no, valiéndose de una función para decidirlo. Una dirección se considerará válida si contiene el símbolo "@".

`````python
def introducir_email(email):
    if "@" in email:
        return True
    else:
        return False

email = input("Introduczca su email: ")

if introducir_email(email):
    print("Email valido")
else:
    print("Email invalido")
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def comprobar_dni(dni):
    palabra = "TRWAGMYFPDXBNJZSQVHLCKE"
    
    if len(dni) < 8 or len(dni) > 9:
        return False

    numero = dni[:-1]
    letra = dni[-1].upper()

    if not numero.isdigit():
        return False

    letra_correcta = palabra[int(numero) % 23]

    return letra == letra_correcta

dni = input("Introduzca su DNI: ")

if comprobar_dni(dni):
    print("DNI valido")
else:
    print("DNI invalido")
`````
