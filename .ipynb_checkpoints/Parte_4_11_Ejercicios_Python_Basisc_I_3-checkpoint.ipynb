## Ejercicio 1
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4
B = "Text"
C = 4.1

print("A == B no son equivalentes al ser un int y un str")
print(A != C)
print(A > C)
print(C <= A)
print(B != C)
````

## Ejercicio 2
Crea un programa donde se recojan dos inputs del usuario, y el output del programa sea si esos inputs son iguales o no

````python
valor1 = input("El valor 1 es: ")
valor2 = input("El valor 2 es: ")
valor1 == valor2
valor1 != valor2
print("Los valores son iguales?", valor1 == valor2)
````

## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
valor1 = input("El valor 1 es: ")
valor2 = input("El valor 2 es: ")
valor3 = input("El valor 3 es: ")
valor1 == valor2 == valor3
print("¿Todos los valores son iguales?", valor1 == valor2 == valor3)
valor1 == valor2
valor2 == valor3
valor3 == valor1
print("¿Al menos dos valores son iguales?", (valor1 == valor2) or (valor2 == valor3) or (valor3 == valor1))
````

## Ejercicio 4
Razona sin ejecutar código el output que obtendremos de las siguientes sentencias
1. True and True and False
2. not ((True or False) and (True or False))
3. (False or False or False or True) and False
4. not False and True and not True

# 1. Al ser and, requiere que sean todos True, al haber un False el resultado es False
# 2. Trabajamos primero con los parentesis que al ser or, cada uno seria True. Luego al ser and tienen que ser iguales y como lo son, seria True. Y por ultimo al ser not al principio, hay que invertir el True a False, por lo tanto el resultado final es False.
# 3. Trabajamos primero con los parentesis que es or con 3 False y 1 True, por lo tanto termina siendo True. Y como and tiene que ser igual a False pero es True, el resultado final es False.
# 4. El primero al ser not False seria True, que al ser and con otro True seria nuevamente True. Por ultimo not True quedaria como False que con and y el anterior True quedaria como resultado final False.
