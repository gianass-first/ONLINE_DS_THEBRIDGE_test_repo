```python
# Ejercicio 1

nombre = input("Introduzca su nombre: ")
nombre = nombre.upper()
print(nombre)
```

```python
# Ejercicio 2

palabra = input("Introduzca cualquier palabra: ")
if 'a' in palabra:
    print(palabra, "contiene 'a'")
else:
    print(palabra, "no contiene 'a'")
```

```python
# Ejercicio 3

palabra = input("Introduzca una palabra: ")
letra = input("Introduzca una letra: ")
if letra in palabra:
    print(palabra, "contiene", letra)
else:
    print(palabra, "no contiene", letra)
```

```python
# Ejercicio 4

palabra = input("Introduzca una palabra: ")
letra = palabra[0]
letra = letra.lower()
if letra in "aeiou":
    print(palabra, "comienza con vocal")
else:
    print(palabra, "no comienza con vocal")
```

```python
# Ejercicio 5

email = input("Introduzca su email: ")
if "@" in email and "." in email:
    print("Email correcto")
else:
    print("Email incorrecto")
```

```python
# Ejercicio 6 Solicita dos palabras y comprueba si son iguales ignorando may√∫sculas.

palabra1 = input("Introduzca la primera palabra: ")
palabra2 = input("Introduzca la segunda palabra: ")

palabra1 = palabra1.lower()
palabra2 = palabra2.lower()

if palabra1 == palabra2:
    print("Las palabras son iguales")
else:
    print("Las palabras no son iguales")
```

```python
# Ejercicio 7 Solicita una palabra y muestra los √∫ltimos 3 caracteres.

palabra = input("Introduzca una palabra: ")
ultimos3caracteres = palabra[-3:]
print(ultimos3caracteres)
```

```python
# Ejercicio 8 Comprueba si una palabra es un pal√≠ndromo.

palabra = input("Introduzca una palabra: ")
if palabra == palabra[::-1]:
    print("La palabra", palabra, "se lee igual al derecho y al reves")
else:
    print("La palabra", palabra, "no se lee igual al derecho y al reves")
```

```python
# Ejercicio 9 Crea una funci√≥n que valide un email (debe contener "@").

email = input("Introduzca un email: ")
def recibir_email(email):
    if "@" in email:
        return True
    else:
        return False

resultado = recibir_email(email)
if resultado:
    print("Email correcto")
else:
    print("Email incorrecto")
```

```python
# Ejercicio 10 Crea una funci√≥n que valide un DNI (n√∫mero + letra correcta).

def recibir_dni(dni):
    if len(dni) != 9:
        return False

    numero = dni[:-1]
    letra = dni[-1]
    
    if not numero.isdigit() or not letra.isalpha():
        return False
        
    numero_int = int(numero)
    letras = "TRWAGMYFPDXBNJZSQVHLCKE"
    letra_correcta = letras[numero_int % 23]

    if letra.upper() == letra_correcta:
        return True
    else:
        return False

dni = input("Introduce tu DNI: ")
if recibir_dni(dni):
    print("DNI v√°lido")
else:
    print("DNI incorrecto")
```
